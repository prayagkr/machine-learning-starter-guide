```python
import pandas as pd
import numpy as np
import datetime
```

```python
today = datetime.datetime.today()
```

```python
df_date_list = [today - datetime.timedelta(days=x) for x in range(0,20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
df['day_of_week'] = df.day.dt.day_name()
df.head(7)
```

```python
weekday_map = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}
```

```python
df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head()
```
